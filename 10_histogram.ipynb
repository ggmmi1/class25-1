```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("images/Unknown.jpeg")
#channel:0-blue, 1 -Green, 2 - Red
hist = cv2.calcHist([img], [0], None, [256], [0,256])

plt.plot(hist, color="b")
plt.title("Histogram")
plt.xlabel("Pixel value")
plt.ylabel("Freqeuncy")
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("images/Unknown.jpeg")
#channel:0-blue, 1 -Green, 2 - Red
hist_1 = cv2.calcHist([img], [0], None, [256], [0,256])
hist_2 = cv2.calcHist([img], [1], None, [256], [0,256])
hist_3 = cv2.calcHist([img], [2], None, [256], [0,256])

plt.plot(hist_1,color="b")
plt.plot(hist_2,color="g")
plt.plot(hist_3,color="r")
plt.title("Histogram")
plt.xlabel("Pixel value")
plt.ylabel("Freqeuncy")
plt.show()
```
